```python
import pandas as pd

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# View first few rows
print(df.head())
print('columns',df.columns)
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop missing rows (if few)
df = df.dropna(subset=['listed_in', 'release_year', 'rating'])


# Convert Released_Year to integer (if needed)
df['release_year'] = df['release_year'].astype(int)
```

```python
from collections import Counter

genres = df['listed_in'].str.split(', ')
genre_list = [g for sublist in genres for g in sublist]
genre_counts = Counter(genre_list)

genre_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)
print(genre_df.head())
```

```python
# Remove duplicates and check for missing values
df.drop_duplicates(inplace=True)
print(df.isnull().sum())

# Clean and convert date_added column
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce', format='mixed')

# Convert release_year to integer (if not already)
df['release_year'] = df['release_year'].astype(int)

# Check results
print(df['date_added'].head())
print(df['release_year'].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x='Count', y='Genre', data=genre_df.head(10))
plt.title('Top 10 Most Common Genres on Netflix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x='release_year', y='rating', data=df, alpha=0.6)
plt.title('Relationship Between Release Year and Content Rating on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Rating')
plt.show()
```

```python
# Count how many shows/movies each rating type has per release year
yearly = df.groupby(['release_year', 'rating']).size().reset_index(name='count')

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(x='release_year', y='count', hue='rating', data=yearly)
plt.title('Netflix Content Ratings Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Rating Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
